```python
# Python Operators - All Types

a = 10
b = 3

# 1️⃣ Arithmetic Operators: +, -, *, /, %, //, **
print("Arithmetic Operators:")
print("a + b =", a + b)   # Addition
print("a - b =", a - b)   # Subtraction
print("a * b =", a * b)   # Multiplication
print("a / b =", a / b)   # Division (float)
print("a // b =", a // b) # Floor division
print("a % b =", a % b)   # Modulus
print("a ** b =", a ** b) # Exponentiation
print()

# 2️⃣ Comparison Operators: ==, !=, >, <, >=, <=
print("Comparison Operators:")
print("a == b:", a == b)  # Equal
print("a != b:", a != b)  # Not equal
print("a > b:", a > b)    # Greater than
print("a < b:", a < b)    # Less than
print("a >= b:", a >= b)  # Greater or equal
print("a <= b:", a <= b)  # Less or equal
print()

# 3️⃣ Logical Operators: and, or, not
print("Logical Operators:")
x = True
y = False
print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)
print()

# 4️⃣ Assignment Operators: =, +=, -=, *=, /=, //=, %=, **=
print("Assignment Operators:")
c = 5
print("c =", c)
c += 2
print("c += 2 ->", c)
c -= 1
print("c -= 1 ->", c)
c *= 3
print("c *= 3 ->", c)
c /= 2
print("c /= 2 ->", c)
c //= 2
print("c //= 2 ->", c)
c %= 2
print("c %= 2 ->", c)
c **= 3
print("c **= 3 ->", c)
print()

# 5️⃣ Bitwise Operators: &, |, ^, ~, <<, >>
print("Bitwise Operators:")
x = 10  # 1010
y = 4   # 0100
print("x & y =", x & y)   # AND
print("x | y =", x | y)   # OR
print("x ^ y =", x ^ y)   # XOR
print("~x =", ~x)         # NOT
print("x << 1 =", x << 1) # Left shift
print("x >> 1 =", x >> 1) # Right shift
print()

# 6️⃣ Membership Operators: in, not in
print("Membership Operators:")
lst = [1, 2, 3, 4, 5]
print("3 in list:", 3 in lst)
print("6 not in list:", 6 not in lst)
print()

# 7️⃣ Identity Operators: is, is not
print("Identity Operators:")
x = [1, 2, 3]
y = x
z = [1, 2, 3]
print("x is y:", x is y)       # True, same object
print("x is z:", x is z)       # False, different objects
print("x is not z:", x is not z)
print()

# 8️⃣ Ternary/Conditional Operator (Python one-liner)
print("Ternary Operator:")
a, b = 10, 20
max_val = a if a > b else b
print("Max value:", max_val)
```

```python
# Python Type Conversion Examples

# Original variables
a = 10        # int
b = 3.14      # float
c = "25"      # string
d = "3.14"    # string representing float
e = True      # boolean

print("Original types:")
print("a:", a, type(a))
print("b:", b, type(b))
print("c:", c, type(c))
print("d:", d, type(d))
print("e:", e, type(e))
print()

# 1️⃣ Integer conversions
print("Integer Conversions:")
print("int(b) =", int(b))       # float -> int (truncates decimal)
print("int(c) =", int(c))       # string -> int
print("int(e) =", int(e))       # True -> 1, False -> 0
print()

# 2️⃣ Float conversions
print("Float Conversions:")
print("float(a) =", float(a))   # int -> float
print("float(c) =", float(c))   # string -> float
print("float(e) =", float(e))   # True -> 1.0
print()

# 3️⃣ String conversions
print("String Conversions:")
print("str(a) =", str(a))       # int -> str
print("str(b) =", str(b))       # float -> str
print("str(e) =", str(e))       # bool -> str
print()

# 4️⃣ Boolean conversions
print("Boolean Conversions:")
print("bool(0) =", bool(0))     # 0 -> False
print("bool(10) =", bool(10))   # non-zero -> True
print("bool('') =", bool(''))   # empty string -> False
print("bool('Python') =", bool('Python'))  # non-empty string -> True
print()

# 5️⃣ Edge Cases
# Converting invalid string to int/float will raise error
invalid_str = "abc"
try:
    int(invalid_str)
except ValueError:
    print("Cannot convert 'abc' to int")
    
try:
    float(invalid_str)
except ValueError:
    print("Cannot convert 'abc' to float")
```

```python
# Python Conditionals - if, elif, else

# 1️⃣ Simple if-elif-else
num = int(input("Enter a number: "))
if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero")
print()

# 2️⃣ Nested if
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
if x > 0:
    if y > 0:
        print("Both numbers are positive")
    else:
        print("x is positive, y is not positive")
else:
    print("x is not positive")
print()

# 3️⃣ Multiple conditions with logical operators
a = int(input("Enter a number: "))
if a > 0 and a % 2 == 0:
    print("Positive and even")
elif a > 0 and a % 2 != 0:
    print("Positive and odd")
elif a < 0:
    print("Negative number")
else:
    print("Zero")
print()

# 4️⃣ Ternary/Conditional operator
age = int(input("Enter your age: "))
status = "Adult" if age >= 18 else "Minor"
print(f"You are: {status}")
print()

# 5️⃣ Edge Cases
# - Input = 0
# - Negative numbers
# - Odd/even check for edge 0 and 1
nums = [0, 1, -1, 2, -5]
for n in nums:
    if n > 0:
        parity = "Even" if n % 2 == 0 else "Odd"
        print(f"{n} is positive and {parity}")
    elif n < 0:
        print(f"{n} is negative")
    else:
        print("Zero encountered")
```

```python
# Python For Loops & Range Function

# 1️⃣ Basic for loop using range()
print("Numbers 0 to 4:")
for i in range(5):  # 0 to 4
    print(i, end=" ")
print("\n")

# 2️⃣ Range with start and end
print("Numbers 1 to 5:")
for i in range(1, 6):  # 1 to 5
    print(i, end=" ")
print("\n")

# 3️⃣ Range with step
print("Even numbers 2 to 10:")
for i in range(2, 11, 2):  # start=2, end=10, step=2
    print(i, end=" ")
print("\n")

# 4️⃣ Reverse loop
print("Numbers 5 to 1 (reverse):")
for i in range(5, 0, -1):  # start=5, end=1 (exclusive), step=-1
    print(i, end=" ")
print("\n")

# 5️⃣ Iterating over list
fruits = ["apple", "banana", "cherry"]
print("Fruits in list:")
for fruit in fruits:
    print(fruit)
print("\n")

# 6️⃣ Nested for loops
print("Multiplication Table 1-3:")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i*j}")
    print()
    
# 7️⃣ Using break in loop
print("Break example:")
for i in range(1, 6):
    if i == 3:
        print("Breaking at 3")
        break
    print(i)
print("\n")

# 8️⃣ Using continue in loop
print("Continue example:")
for i in range(1, 6):
    if i == 3:
        print("Skipping 3")
        continue
    print(i)
print("\n")

# 9️⃣ Edge Cases
# - Empty range
print("Empty range example:")
for i in range(5, 1):
    print(i)  # Will not print anything

# - Negative steps
print("Negative step example:")
for i in range(10, 5, -2):
    print(i)
```

```python
# Python While Loops

# 1️⃣ Basic while loop
print("Basic while loop from 1 to 5:")
i = 1
while i <= 5:
    print(i, end=" ")
    i += 1
print("\n")

# 2️⃣ While loop with break
print("While loop with break:")
i = 1
while i <= 5:
    if i == 3:
        print("Breaking at 3")
        break
    print(i)
    i += 1
print("\n")

# 3️⃣ While loop with continue
print("While loop with continue:")
i = 0
while i < 5:
    i += 1
    if i == 3:
        print("Skipping 3")
        continue
    print(i)
print("\n")

# 4️⃣ Nested while loops
print("Nested while loop - multiplication table 1-3:")
i = 1
while i <= 3:
    j = 1
    while j <= 3:
        print(f"{i} x {j} = {i*j}")
        j += 1
    i += 1
    print()
    
# 5️⃣ Infinite loop handling
# Warning: The following loop is commented because it will run forever
# while True:
#     print("This will run infinitely until break is used")

# 6️⃣ Edge cases
# - Zero or negative range handled manually
count = -3
while count < 0:
    print("Negative start:", count)
    count += 1
print("\n")

# - While loop with empty condition
n = 0
while n < 0:
    print("This will not print anything because condition is False")
```
